```python
def check_season(month):
    if month in ['December', 'January', 'February']:
        print('Winter')
    elif month in ['March', 'April', 'May']:
        print('Spring')
    elif month in ['June', 'July', 'August']:
        print('Summer')
    print('Autumn')

check_season('October')
```

```python
def add_item(lists, item):
    lists.append(item)
    return lists


names = ['fortune', 'favour', 'victory']
add_item(names, 'sarah')
```

```python
def remove_item(lists, item):
    lists.remove(item)
    return lists

names = ['fortune', 'favour', 'victory', 'sarah']

remove_item(names, 'sarah')
```

```python
def sum_numbers(n):
    lists = []
    for i in range(n+1):
        lists.append(i)
    a = sum(lists)
    return a

sum_numbers(3)
```

```python
def sum_odds(n):
    total = 0
    for i in range(n):
        if i%2 != 0:
            total += i
    return total

sum_odds(7)
```

```python
def sum_evens(n):
    total = 0
    for i in range(n):
        if i%2 == 0:
            total += i
    return total

sum_evens(7)
```

```python
def evens_and_odds(n):
    t_even = 0
    t_odd = 0
    for i in range(n):
        if i%2 == 0:
            t_even += 1
        else:
            t_odd += 1
    return t_even, t_odd
    

evens_and_odds(7)
```

```python
def factorial(n):
    list1 = 1
    for i in range(n+1):
        if i == 0:
            continue
        list1 *= i
    return list1

factorial(4)

```
