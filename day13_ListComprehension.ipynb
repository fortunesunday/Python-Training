```python
numbers = [-4, -3, -2, -1, 0, 2, 4, 6]
negative_and_zero = [x for x in numbers if x<= 0]
negative_and_zero
```

```python
list_of_lists =[[[1, 2, 3]], [[4, 5, 6]], [[7, 8, 9]]]
flattened_list = [i for d in list_of_lists for  r in d for i in r]
flattened_list
```

```python
countries = [[('Finland', 'Helsinki')], [('Sweden', 'Stockholm')], [('Norway', 'Oslo')]]
flattened_country = [country for row in countries for tuples in row for country in tuples]
flattened_country
```

```python
names = [[('Asabeneh', 'Yetayeh')], [('David', 'Smith')], [('Donald', 'Trump')], [('Bill', 'Gates')]]
name_list = [name for list in names for tuples in list for name in tuples]
name_list
```
